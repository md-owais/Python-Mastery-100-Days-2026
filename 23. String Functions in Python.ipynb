# 5. format

```python
"Hello my name is {} and I am {}".format("Owais", 25)
```

```python
"Hello my name is {1} and I am {0}".format("Owais", 25)
```

```python
"Hello my name is {0} and I am {1}".format("Owais", 25)
```

```python
"Hello my name is {name} and I am {age}".format(name = "Owais", age= 25)
```
