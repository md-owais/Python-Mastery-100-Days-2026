# Recursion using Python

```python
# Function call itself
# 4*5 -> repeated addition - 4+4+4+4+4 = 20

# Using Loop
def multiply(a,b):
    result = 0
    
    for i in range(b):
        result = result + a
        
    print(result)
multiply(3,4)
```

```python
# Using recursion
def mul(a,b):
    
    if b==1:
        return a
    else:
        return a + mul(a,b-1)
print(mul(5,6))
```

```python
def fact(number):
    if number == 1:
        return 1
    else:
        # 5 = 5*4!
        return number * fact(number - 1)
print(fact(5))
```

```python
# palindrome
# step 1 - base condition - a - one char is palindrome
def palin(text):
    if len(text) <= 1:
        print("Palindrome")
    else:
        if text[0] == text[-1]:
            palin(text[1:-1])
        else:
            print("Not a Palindrome")

palin("madam")
palin("malayalam")
palin("Python")
```

```python
palin("abba")
```

# The Rabbit Problem
If 2 newborn rabbits are put in a pen, how many rabbits will be in the pen after 1 year?

Assume that rabbits...

..always produce one male and one female offspring

..can reproduce once every month

...can reproduce once they are one month old

...never die!

```python
# Fibonacci
def fib(m):
    if m == 0 or m == 1:
        return 1
    else:
        return fib(m-1) + fib(m-2)
print(fib(12))    
```

```python
import time
def fib(m):
    if m == 0 or m == 1:
        return 1
    else:
        return fib(m-1) + fib(m-2)

start = time.time()    
print(fib(24))
print(time.time() - start)
```
