# Operators
Operators are used to perform operations on variable and values. Python has the following operators:
1. Arithmetic Operators
2. Comparison Operators
3. Logical Operators
4. Bitwise Operators
5. Assignment Operators
6. Identity Operators
7. Membership Operators

```python
x = 5
y = 2

print(x + y)
```

```python
print(x - y)
```

```python
print(x * y)
```

```python
print(x / y)
```

```python
print(x % y)
```

```python
print(x ** y)
```

```python
print(x // y)
```

```python
# Comparison Operators
print(x > y)
```

```python
print(x < y)
```

```python
print(x >= y)
```

```python
print(x <= y)
```

```python
print(x == y)
```

```python
print(x != y)
```

```python
# Logical Operators
x = True
y = False
print(x or y)
```

```python
print(x and y)
```

```python
print(not x)
```

```python
print(not y)
```

```python
# Bitwise Operators - works on Binary values
x = 2 # 2 - 010 
y = 3 # 3 - 110 - final after & - 010

print(x & y)
```

```python
print(x | y) # 110
```

```python
print(x >> 2)
```

```python
print(y << 3)
```

```python
print(~x)
```

```python
# Assignment Operators
a = 3
print(a)
```

```python
a += 3 # a = a + 3
print(a)
```

```python
a++ # not support in Python
++a
```

```python
# Identity Operator - two variables are available at the same location or not
a = 3
b = 3
print(a is b)
```

```python
a = "Hello"
b = "Hello"

print(a is b)
```

```python
a = [1,2,3] 
b = [1,2,3] 
print(a is b)
```

```python
a = "Hello-world"
b = "Hello-world"
print(a is b)
```

```python
a = "Hello-world"
b = "Hello-world"
print(a is not b)
```

```python
# Membersip operators - let you whether it's available inside another
x = 'Delhi'
print("D" in x)
```

```python
x = 'Delhi'
print("D" not in x)
```

```python
x = [1,2,3]
print(5 in x)
```

```python
x = (1,2,3)
print(5 in x)
```
