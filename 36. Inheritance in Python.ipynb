```python
# Inheritance - gives code reusability
# Private Members are not inherited

class User: # parent class
    
    def login(self):
        print("Login")
        
    def register(self):
        print("Register")
        
class Student(User):  #child class
    
    def enroll(self):
        print("Enroll")
        
    def review(self):
        print("Review")
        
stu1 = Student()

stu1.enroll()
stu1.review()
stu1.login()
stu1.register()
```
