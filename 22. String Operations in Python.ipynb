```python
# Loop
c = 'hello world'
for i in c:
    print(i)
```

```python
c = 'hello world'
for i in c[2:7:2]:
    print(i)
```

```python
c = 'hello world'
for i in c[::-1]:
    print(i)
```

```python
# Membership Operators
'h' in c
```

```python
'H' in c
```
