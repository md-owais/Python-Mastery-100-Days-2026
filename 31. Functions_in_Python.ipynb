```python
# The 2 Point of Views
def is_even(number):
    if type(number) == int:
        '''
        This function tells if given no is odd or even
        Input - any valid integer
        output - odd/even
        created By - Owais
        Last edited - 25th March
        '''
        if number%2==0:
            return "Even"
        else:
            return "Odd"
    else:
        return "Not allowed"

for i in range(1, 11):
    print(is_even('Hello'))
```
