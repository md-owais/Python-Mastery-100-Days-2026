# 12. Write a program to find the volume of the cylinder. Also find the cost, when the cost of 1litre milk is 40Rs.
Cylinder:

Volume = pi*r^2*h

cost = volume * cost per unit

```python
r = float(input("Enter the radius: "))
h = float(input("Enter the height: "))

volume = 3.14 * r * r *h

print("The Volume of Cylinder is:", volume)
```

```python
r = float(input("Enter the radius: "))
h = float(input("Enter the height: "))
cost_per_litre_milk = float(input("Enter the cost per liter milk: "))

volume = 3.14 * r * r *h

print("The Volume of Cylinder is:", volume)

cost = volume * cost_per_litre_milk

print("Cost per litre Milk is:", cost)
```
