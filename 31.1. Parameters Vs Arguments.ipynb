```python
def power(a,b):
    return a ** b
power(3)
```

```python
def power(a,b):
    return a ** b
power()
```

```python
# default Arg
def power(a=1,b=1):
    return a ** b
power(2,3)
```

```python
def power(a=1,b=1):
    return a ** b
power(2)
```

```python
power()
```

```python
# Positional Arguments - In which order passing arguments and getting assigned in same way
power(2,3)
```

```python
# The below behavior is positional arguments
def power(a=1,b=1):
    return a ** b
power(2,3)
```

```python
# Keyword Arguments - -> then also it will will take a vaule only first and it's over write
power(b=2, a= 3)
```

```python
print(1,2,3)
```

```python
# Arbitary function
def flexi(*number):
    prodcut = 1
    for i in number:
        prodcut = prodcut * i
    print(prodcut)
```

```python
flexi(1)
```

```python
flexi(1,2)
```

```python
flexi(1,2,3)
```

```python
flexi(1,2,3,4,5,6,7)
```
