# Editing and Deleting in Strings

```python
c = "Hello"
c[0] = 'X'
```

```python
# Note: Strings are a Immutable Data Types
c = "World"
c
```

```python
c[5] = 'X'
```

```python
del c
```

```python
print(c)
```

```python
c = 'hello'
del c[0]
```

```python
del c[:3:2]
```
