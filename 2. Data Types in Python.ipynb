Python supports 3 categories of data types
. Basic Types - (Integer, float, complex, boolean, and string)
. Container Types - (List, Tuples, Sets and Dictionaries)
. User-defined types - Class

```python
# Integer
print(4)
print(1e308) # 10^308 (10 to the power 308 digit excepting)
print(1e309) # infinity
```

```python
# float
print(4.5)
print(1.7e308)
print(1.7e309)
```

```python
# boolean
print(True)
print(False)
```

```python
# complex
print(4+5j)
```

```python
# string
print('Bangalore')
print("Bihar")
print("""Bangalore""")
```

```python
# List
print([1,2,3,4,5])
```

```python
# Tuple
print((1,2,3,4,5))
```

```python
# sets
print({1,2,3,4,5})
```

```python
# dict
print({"Name":"Owais","Age":30,"Gender":"Male"})
```
