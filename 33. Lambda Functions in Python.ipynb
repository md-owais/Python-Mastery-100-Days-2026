```python
# Lambda functions also known as anonymous functions
# syntax: lambda input: expression
x = lambda x:x**2
x(9)
```

```python
a = lambda x, y: x + y
a(4,5)
```

```python
# why lambda function used?
# Along with Higher order functions
b = lambda x:x[0] == 'a'
b('apple')
```

```python
b('banana')
```

```python
b = lambda x:"Even" if x%2 == 0 else "Odd"
b(3)
```

```python
# Higher Order Function
def return_sum(func, L):
    result = 0
    for i in L:
        if func(i):
            result = result + i
    return result
    
L = [11, 14, 21, 23, 56, 78, 45, 29, 28]

x = lambda x: x % 2 == 0
y = lambda x: x % 2 != 0
z = lambda x: x % 3 == 0

print(return_sum(x, L))
print(return_sum(y, L))
print(return_sum(z, L))
```

```python
# Map Functions
L = [1,2,3,4,5,6,7]
L
```

```python
map(lambda x:x*2, L)
```

```python
list(map(lambda x:x*2, L))
```

```python
list(map(lambda x: x%2 == 0, L))
```
