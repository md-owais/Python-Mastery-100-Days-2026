# Accessing Substrings from a String

```python
# Concept of Indexing
c = "hello"
print(c)
```

```python
print(c[0])
```

```python
print(c[5])
```

```python
# Types of Indexing - Positive Indexing, Negative Indexing
print(c[-1])
```

```python
print(c[-2])
```

```python
print(c[-6])
```

```python
# Slicing
c = "hello world"
print(c[0:5])
```

```python
print(c[2:])
```

```python
print(c[:4])
```

```python
print(c[:])
```

```python
print(c[2:6:2])
```

```python
print(c[0:8:3])
```

```python
print(c[0:6:-1]) # Positive indexing with negative won't give any output
```

```python
print(c[-5:-1:2])
```

```python
print(c[::-1]) # : ->starting and : -> ending and after that reversing
```

```python
print(c[-1:-5:-1])
```
