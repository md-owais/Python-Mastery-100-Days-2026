# 1. Single Inheritance

A child class inherits from one parent class.

```python
class Parent:
    def show(self):
        print("This is Parent class")

class Child(Parent):
    def display(self):
        print("This is Child class")

obj = Child()
obj.show()
obj.display()
```

# 2. Multiple Inheritance

A child class inherits from multiple parent classes.

```python
class Father:
    def skill1(self):
        print("Gardening")

class Mother:
    def skill2(self):
        print("Cooking")

class Child(Father, Mother):
    def skills(self):
        print("Child skills")

obj = Child()
obj.skill1()
obj.skill2()
obj.skills()
```

# 3. Multilevel Inheritance

A class inherits from a parent, and another class inherits from that child (chain).

```python
class Grandparent:
    def home(self):
        print("Grandparent's home")

class Parent(Grandparent):
    def car(self):
        print("Parent's car")

class Child(Parent):
    def bike(self):
        print("Child's bike")

obj = Child()
obj.home()
obj.car()
obj.bike()
```

# 4. Hierarchical Inheritance

Multiple child classes inherit from a single parent class.

```python
class Parent:
    def property(self):
        print("Parent property")

class Child1(Parent):
    def show1(self):
        print("Child1")

class Child2(Parent):
    def show2(self):
        print("Child2")

obj1 = Child1()
obj2 = Child2()

obj1.property()
obj2.property()
```

# 5. Hybrid Inheritance

Combination of two or more types of inheritance.

```python
class A:
    def methodA(self):
        print("Class A")

class B(A):
    def methodB(self):
        print("Class B")

class C(A):
    def methodC(self):
        print("Class C")

class D(B, C):
    def methodD(self):
        print("Class D")

obj = D()
obj.methodA()
obj.methodB()
obj.methodC()
obj.methodD()
```
