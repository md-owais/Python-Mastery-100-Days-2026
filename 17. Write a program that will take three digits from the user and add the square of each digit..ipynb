# 17.	Write a program that will take three digits from the user and add the square of each digit.

```python
num = int(input("Enter three Digit number: "))
a = num % 10 # Last digit
num = num // 10

b = num % 10 # Middle Digit
c = num // 10 # First Digit

# result = a*a + b*b + c*c
result = (a**2) + (b**2) + (c**2)
print("The sum of three Digit is:", result)
```

```python
num = input("Enter a 3-digit number: ")

result = int(num[0])**2 + int(num[1])**2 + int(num[2])**2

print("Sum of squares of digits:", result)
```

```python
num = input("Enter a 3-digit number: ")

result = 0
for digit in num:
    result += int(digit) ** 2
print("Sum of squares:", result)
```
