# Generators in Python

Python generators are a simple way of creating iterators

```python
def gen_demo():
    yield "First statement"
    yield "Second statement"
    yield "Third statement"
    
gen = gen_demo()
print(gen)
```

```python
print(next(gen))
```

```python
print(next(gen))
```

```python
print(next(gen))
```

```python
# In function there's return function but for generator has similar yield
gen = gen_demo()
for i in gen:
    print(i)
```

# Difference between yield and return

```python
# return function - exits after executing the function but generator partially remembers the previous executed functions
# Example 2

def square(num):
    for i in range(1, num+1):
        yield i**2
gen = square(10)       
print(next(gen))
```

```python
print(next(gen))
```

```python
print(next(gen))
```

```python
for i in gen:
    print(i)
```

# Generator Expression

```python
L = [i**2 for i in range(1,101)]
gen = (i**2 for i in range(1,101))

for i in gen:
    print(i)
```

# Memory effiecient

```python
L = [x for x in range(10000)]
gen = (x for x in range(10000))

import sys

print("Size of L in memory:", sys.getsizeof(L))
print("Size of L in memory:", sys.getsizeof(gen))
```

# Representing Infinite Streams

```python
def all_even():
    n = 0
    while True:
        yield n
        n += 2
        
even_num_gen = all_even()
next(even_num_gen)
next(even_num_gen)
```

# Chaining Generators

```python
def fibonacci_number(nums):
    x, y = 0, 1
    for _ in range(nums):
        x, y = y, x+y
        yield x
        
def square(nums):
    for num in nums:
        yield num **2
print(sum(square(fibonacci_number(10))))
```
